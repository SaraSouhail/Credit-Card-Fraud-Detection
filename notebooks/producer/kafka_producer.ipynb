```python
from kafka import KafkaProducer
import pandas as pd
import json
import time

# Charger le CSV sans Class
df = pd.read_csv("../../Big_Data/mock_credit_card_transactions_no_class.csv")
# df = pd.read_csv("../../test_transactions_no_class/Data_no_class.csv")

# Créer le producer Kafka
producer = KafkaProducer(
    bootstrap_servers="localhost:9092",
    value_serializer=lambda v: json.dumps(v).encode("utf-8")
)

topic_name = "credit-card-transactions"

# Envoyer chaque ligne comme un message Kafka
for _, row in df.iterrows():
    message = row.to_dict()
    producer.send(topic_name, value=message)
    
    time.sleep(0.5)  # simulation streaming temps réel

producer.flush()
producer.close()
```
