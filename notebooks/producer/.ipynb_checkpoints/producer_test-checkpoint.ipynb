```python
import pandas as pd

# Load original file WITH Class
df = pd.read_csv("../../test_transactions/data_test.csv")

# Count frauds
total_rows = len(df)
fraud_count = (df["Class"] == 1).sum()
non_fraud_count = (df["Class"] == 0).sum()

print("Total rows:", total_rows)
print("Fraud (Class = 1):", fraud_count)
print("Non-Fraud (Class = 0):", non_fraud_count)
print("Fraud ratio:", fraud_count / total_rows)
```

```python
# 
from kafka import KafkaProducer
import pandas as pd
import json
import time

# -------------------------
# 1️⃣ Read CSV
# -------------------------
df = pd.read_csv("../../test_transactions_no_class/Data_no_class.csv")

# 🔴 Make sure columns are numeric
df = df.astype(float)

# -------------------------
# 2️⃣ Kafka Producer
# -------------------------
producer = KafkaProducer(
    bootstrap_servers="localhost:9092"
)

topic = "topic"   # 🔴 MUST match consumer

# -------------------------
# 3️⃣ Send rows one by one
# -------------------------
for i, row in df.iterrows():
    message = row.to_dict()

    producer.send(
        topic,
        value=json.dumps(message).encode("utf-8")
    )
    
    # print(f"Sent row {i}")
    time.sleep(0.1)   # simulate real-time stream

producer.flush()
producer.close()

print("✅ All rows sent")
```
